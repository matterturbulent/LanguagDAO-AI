```python
# open the file for reading
with open('peopledao_snapshot_1.19.22.json', 'r', encoding='utf-8') as file:
     json_string = file.read()

# Use json.loads to parse the JSON string
json_data = json.loads(json_string)
df = pd.json_normalize(json_data['proposals'])

# indlude only official governance proposals by filtering for "PIP"
df = df[(df.title.str.contains("PIP"))]

# Loop through each "body" text of the proposals,
# filtering for whether the proposal has chinese written in it
# if yes, split the english and chinese into two columns of a new dataframe
chinese_translation = []
english_translation = []
for body in df["body"]:
    for i in range(len(body)):
        if ord(body[i]) >= 0x4E00 and ord(body[i]) <= 0x9FCC:
            english_translation.append(body[:i])
            chinese_translation.append(body[i:])
            break
        else:
            continue
english_series = pd.DataFrame({"english":english_translation})
chinese_series = pd.DataFrame({"chinese":chinese_translation})
training_data_addition = pd.concat([english_series, chinese_series], axis=1)

# reverse order of rows 
training_data_addition = training_data_addition.loc[::-1]

# add the new dataframe to the end of the existing training data
training_data_addition.rename(columns={"english": "prompt", "chinese":"completion"}, inplace = True)
old_data = pd.read_csv('training_data.csv', encoding='utf-8')
new_training = pd.concat([old_data, training_data_addition], ignore_index=True)
new_training.to_csv('training_data.csv', index=False, encoding='utf-8')

display(new_training)
```
